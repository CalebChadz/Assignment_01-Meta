```python
import csv
from PIL import Image, ImageDraw

#Set up global list and drawing elements
boxes = []
im = Image.new('RGB', (1600, 2800), (255, 255, 255))
draw = ImageDraw.Draw(im)

#Method to draw the rectangles on to an image file in the same folder
def drawRect(x, y, width, height):
    draw.rectangle((x,y,width,height), fill=(255, 0 , 0), outline=(0, 0, 0))
    im.save('./image.jpg', quality=95)

#Hard-coded csv file read in to a list
#** Note to Caleb: IDK how to make this a dialog box thing, maybe yours is better
with open('toy_problem.csv', newline = '') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        boxes.append(row)

#Rectangle object class initiating all the properties
class Rectangle:
    def __init__(self, index, width, height):
        self.index = index
        self.xpos = 0
        self.ypos = 0
        self.width = width
        self.height = height
        self.drawn = False

#Solution object class is a representation of a solution, which is stored as a list of rectangles
class Solution:
    def __init__(self, boxes = boxes):
        self.boxlist = []
        self.drawnlist = []
        for i in range(0, len(boxes)):
            rect = Rectangle(i, boxes[i][0], boxes[i][1])
            self.boxlist.append(rect)
     
    #Method to get the height at which a Rectangle can be drawn without overlap at a given point
    def getMaxHeight(self, x, width):
        maxHeight = 0
        for box in self.boxlist:
            if box.xpos < x + width and box.ypos + (40 * int(box.height)) >= maxHeight 
            and box.drawn and x < box.xpos + (40 * int(box.width)):
                maxHeight = box.ypos + (40 * int(box.height))
        return maxHeight
    
    #Method to stack/place the Rectangles onto a sheet
    def stackBoxes(self, x, y, width, height):
        if x + width > im.width:
            x = 0
        y = self.getMaxHeight(x, width)
        drawRect(x, y, x + width, y + height)
        x += width
        return x, y

#Initial solution generator, the Rectangles read in earlier are placed onto the sheet in the order 
#they appear on the CSV
def initialSolution():
    temp_boxes = Solution(boxes)
    init_x = 0
    init_y = 0
    #**Note to Caleb: Iterator cause I was testing, but would be later swapped in for a foreach loop
    for i in range(100):
        box = temp_boxes.boxlist[i]
        box.xpos = init_x
        box.ypos = init_y
        init_x, init_y = temp_boxes.stackBoxes(init_x, init_y, 40 * int(box.width), 40 * int(box.height))
        box.ypos = init_y
        box.xpos = init_x - 40 * int(box.width)
        #print(str(box.xpos) + ' ' + str(box.ypos) + ' ' + str(40 * int(box.width)) + ' ' + str(40 * int(box.height)))
        box.drawn = True
        
initialSolution()
```
